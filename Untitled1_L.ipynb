```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Power-Networks-LCL-June2015(withAcornGps)v2_1.csv", header=0)
```

```python
data.shape
```

```python
data.columns
```

```python
date = pd.to_datetime(data['DateTime'])
```

```python
data.head()
```

```python
data = data.loc[:, ["KWH/hh (per half hour) "]]
data = data.set_index(date)
data["KWH/hh (per half hour) "] = pd.to_numeric(data["KWH/hh (per half hour) "], downcast='float', errors='coerce')
```

```python
weekly = data.resample("W").sum()
weekly.plot(style=[':', '--', '-'])
```

```python
idx = 0
le = []
for i in range(10):
    idx = idx + 48*i
    temp =  data.iloc[idx:idx + 48]
    temp.plot()
    le.append([float(temp.mean()), float(temp.sum()), float(temp.std()), float(temp.max())])
```

```python
le = np.array(le)
le
```

```python
le[:, 3]
```

```python
for i in range(le.shape[1]):
    plt.plot(le[:, i])
    plt.show()
plt.show()
```
